# 2. RKHS Spaces, Mean Embeddings, and Discrepancy Between Distributions

Let $k : \chi \times \chi \to \mathbb{R}$ be a symmetric positive-definite kernel, and let $\mathcal{H}_k$ be the associated reproducing-kernel Hilbert space.



**2.1.** Starting from a finite sample $\{x_n\}_{n=1}^{N}$, obtain the empirical form of the mean embedding in an RKHS. Then show how this quantity can be evaluated through the kernel in two cases: (i) the linear kernel $k(x_i, x_j) = x_i^{\top}x_j$, and (ii) a Gaussian RBF kernel. In your derivation, indicate the dimensions of the objects involved whenever appropriate and briefly compare the interpretation of both kernels.



**2.2.** Explain the difference between a raw moment, a centered moment, and a standardized moment, both in the original space and in the kernel-induced space.



**2.3.** Let $X$ and $Y$ be random variables taking values in $\chi$, distributed according to $P$ and $Q$, respectively, and consider their mean embeddings in $\mathcal{H}_k$,

$$
\mu_P = \mathbb{E}[k(X, \cdot)], \qquad \mu_Q = \mathbb{E}[k(Y, \cdot)].
$$

Starting from these expressions, formulate the discrepancy measure

$$
\lVert \mu_P - \mu_Q \rVert_{\mathcal{H}_k}^{2},
$$

and derive its matrix-form solution using its empirical version based on finite samples from $X$ and $Y$. Then implement this measure in code under the assumption that $X$ contains data from one MNIST class and $Y$ contains data from another class. Interpret the obtained results, indicating which pairs of classes exhibit smaller discrepancy and which exhibit larger discrepancy, and explain the meaning of these results.
