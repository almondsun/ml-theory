# 1. Regression in Feature Spaces and Kernel Formulation

Let an i.i.d. dataset

$$
\{(x_n, t_n)\}_{n=1}^{N}, \qquad x_n \in \mathbb{R}^{P}, \quad t_n \in \mathbb{R},
$$

be given, and let

$$
\phi : \mathbb{R}^{P} \to \mathcal{H}
$$

be a mapping into a feature space $\mathcal{H}$, not necessarily finite-dimensional.

Consider the model

$$
t_n = \langle w, \phi(x_n) \rangle_{\mathcal{H}} + \eta_n, \qquad \eta_n \sim \mathcal{N}(0, \sigma_{\eta}^{2}).
$$



**1.1.** Formulate the regularized least-squares problem in the feature space $\mathcal{H}$. In your answer, explicitly write the objective function to be minimized, indicate the unknown variable of the problem, and briefly explain the role of the regularization parameter.



**1.2.** Starting from the problem formulated in the previous item, obtain its solution. To do so, write the problem in matrix notation, develop the optimality conditions step by step, and arrive at the final expression for the estimator.



**1.3.** Consider the polynomial mapping

$$
\phi(x) = \begin{bmatrix} 1 \\ x \\ x^{2} \end{bmatrix}.
$$

For a dataset $\{(x_n, t_n)\}_{n=1}^{N}$, construct the design matrix associated with this mapping and implement in Python the solution obtained in the previous item. Use a simple numerical example to verify the result and briefly interpret the solution obtained.
