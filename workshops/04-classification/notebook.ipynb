```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.datasets import mnist

(X_train, y_train), (X_test, y_test) = mnist.load_data()

X_train, X_test  = X_train / 255.0, X_test / 255.0
X_train, X_test = X_train.reshape(X_train.shape[0], 28*28), X_test.reshape(X_test.shape[0], 28*28)

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
np.unique(y_train)
```

```python
Input = tf.keras.layers.Input(shape=(28*28,))
layer_1 = tf.keras.layers.Dense(128, activation='relu')(Input)
layer_2 = tf.keras.layers.Dense(128, activation='relu')(layer_1)
layer_3 = tf.keras.layers.Dense(10, activation='softmax')(layer_2)  # Softmax or sigmoid -> These are the only ones that have a probability interpretation. The rest of the activation functions are just a transformation of the input data. The softmax function is used for multi-class classification problems, while the sigmoid function is used for binary classification problems.

model = tf.keras.Model(inputs=Input, outputs=layer_3)

model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=30, batch_size=32, validation_data=(X_test, y_test))
```

```python
# test
model.evaluate(X_test, y_test)
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)
y_pred_classes[:10], y_test[:10]
```

```python
Input = tf.keras.layers.Input(shape=(28, 28))
layer_1 = tf.keras.layer.Conv2D(32, kernel_size=(3, 3), activation='relu')(Input)
layer_2 = tf.keras.layer.Conv2D(64, kernel_size=(3, 3), activation='relu')(layer_1)
layer_3 = tf.keras.layer.Flatten()(layer_2)
layer_4 = tf.keras.layer.Dense(128, activation='relu')(layer_3)
Output = tf.keras.layer.Dense(10, activation='softmax')(layer_4)

model = tf.keras.Model(inputs=Input, outputs=Output)
```
